### Lists: Slicing

- Returns several items from a list
- Syntaxes: use square bracket and colon between two indices.
- <font face="Courier New">list[starting_index:ending_index]</font> returns items in the list from <font face="Courier New">starting_index</font> up to
(but not including) the <font face="Courier New">ending_index</font>.
- When <font face="Courier New">starting_index</font> is not specified (or blank), it returns items from the beginning of the list
- When <font face="Courier New">ending_index</font> is not specified (or blank), it returns items until the last item


```python
color = ["blue","white","purple","pink","green"]
color[1:]
```

```python
color[:3]
```

```python
color[1:-1]
```

### Loops

```python
# Use the built-in function "range"
# to create lists of numbers easily
for i in range(0,4):
    print("{0} squared is {1}".format(i, i**2))
```

### Addition assignment operator

<font face="Courier New" color="blue"><b>+=</b></font> adds two values together and assigns the resultant value to a variable.

<br><font face="Courier New">var1 += var2</font> is the same as <font face="Courier New">var1 = var1 + var2</font>.

<br>The addition assignment operator is often used as a condition in a while loop.

```python
i = 0
while  i < 4:
    print("{0} squared is {1}".format(i, i**2))
    i += 1
```

### Loops: Combining with if-else statement

Addition assignment operator can also be useful to count

```python
my_bags = ["Gucci", "Zara", "H&M", "Louis Vuitton"]
luxury = ["Louis Vuitton", "Chanel", "Dior", "Prada", "Hermes", "Gucci"]
count_expensive = 0
count_less_expensive = 0

for a_bag in my_bags:
    if a_bag in luxury:
        count_expensive += 1
        print("{0} is expensive!".format(a_bag))
    else:
        count_less_expensive += 1
        print("{0} is cool".format(a_bag))

print("I have {0} expensive bags and {1} less expensive bags".format(count_expensive, count_less_expensive))
```

### Exercise 2

Melissa's world trips are cancelled due to COVID-19 pandemic. She receives a total refund of €10,000 in cash. She decides to invest the money in a savings account yielding 5% interest per annum (compounded in arrear). <br>

a) Assuming that Melissa leaves all interest on deposit in the account, calculate and display the amount of money in the account at the end of each year for 10 years. <br>
b) Now assume that Melissa wants to take all the interest accummulated automatically when the amount of money in the account reaches €14,000, leaving only the principal to be accummulated again until the end of year 10. Display the amount of money that she withdraws and recalculate the amount of money in the account again until the end of year 10. <br>

Tip: Use the following formula for determining these amounts: <script src='https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.4/MathJax.js?config=default'></script> $$ {p(1 + r)^n} $$

where <br>
<i>p</i> is the original amount invested (i.e., the principal) <br>
<i>r</i> is the annual interest rate <br>
<i>n</i> is the number of years

```python
# Exercise 2 a)
amount = 0
principal = 10000
rate = 0.05
for year in range(1, 11):
    amount = principal * (1 + rate) ** year
    print(f'{year:2}{amount:10.2f}')
```

```python
# Exercise 2 b)
principal = 10000
threshold = 14000
rate = 0.05
withdrawal = 0
for year in range(1, 11):
    if (year == 1):
        amount = principal
    amount = amount*(1+rate)
    print(f"Amount at the end of year {year:12} is {amount:10.2f}")
    if amount >= threshold:
        withdrawal = amount-principal
        amount = principal
        print(f"Amount withdrawn at the end of year {year:2} is {withdrawal:10.2f}")
```
