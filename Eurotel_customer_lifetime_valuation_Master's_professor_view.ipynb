### How?

**A customer is considered to be an asset to the company.**


* Valuation of that asset:

<script src='https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.4/MathJax.js?config=default'></script> $$ \text{CLV}=\sum_{t=1}^{T}\dfrac{p_t - c_t}{(1+i)^t} - \text{AC}. $$

* <script   src='https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.4/MathJax.js?config=default'></script> $p_t$ = price paid by a customer at time t
* <script src='https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.4/MathJax.js?config=default'></script> $c_t$ = direct cost of serving the customer at time t
* <script src='https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.4/MathJax.js?config=default'></script> $\frac{1}{(1+i)^t}  $ = discount rate or cost of capital of the firm
* AC = acquisition cost
* <script src='https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.4/MathJax.js?config=default'></script> $T$ = time horizon for estimating CLV

### WACC: Equity

The owners’ claims to the assets of a corporation are called shareholders’ equity or simply equity.

<ul>
<li> Share Capital. The Share Capital account shows the owners’ investment in the corporation.
<li> Retained Earnings. The Retained Earnings account shows the cumulative net income earned, minus cumulative net losses and dividends.
<li> Dividends. Dividends are optional; they are declared by the board of directors (may require
shareholders’ approval at the corporation’s annual general meeting).
</ul>

<script src='https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.4/MathJax.js?config=default'></script> $ r_{Equity}  $ (Cost of equity) is the rate of return an investor requires before exploring other opportunities.

### WACC: Debt

Debt is something, usually money, borrowed by one party from another.
<br/><br/>

<script src='https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.4/MathJax.js?config=default'></script> $ r_{Debt} $ (Cost of debt) is the effective interest rate that a company pays on its debts, such as bonds and loans.

### WACC

Weighted Average Cost of Capital.

Measures a company's cost to borrow money.

High WACC: company is less likely to create value.

The formula for WACC equals:

<script src='https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.4/MathJax.js?config=default'></script> $$ WACC=\dfrac{Equity}{Equity + Debt}r_{Equity} +  \dfrac{Debt}{Equity + Debt}r_{Debt} (1- Tax\_rate). $$

Businesses are able to deduct interest expenses from their taxes. Because of this, the net cost of a company’s debt is the amount of interest it is paying minus the amount it has saved in taxes.

In order for a project to be worthwhile, it must return at least the WACC.

### Customer lifetime

Important in any CLV analysis, is the time horizon under consideration. For each customer, we can determine the expected customer relationship duration (CRD) based on that customer's probability of churning:

For each customer j:

<script src='https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.4/MathJax.js?config=default'></script> $$ CRD_j=\dfrac{1}{P(churn)}. $$

Let's calculate the CRD and include this information in a new column `CRD`. We work with rounded values.

```python
df['CRD'] = round(1 / df['ChurnProbabilities'])
df.head(5)
```

### Customer lifetime

Some customers have a very high CRD (and thus a very low churn probability). This means that they are likely to stick with our company for a long time.

Others have a very high churn probability and thus a low CRD. These people have a high probability of switching to a competitor.

Let's take a look at the minimum and maximum CRD in our data.

```python
min_crd = int(min(df['CRD']))
max_crd = int(max(df['CRD']))

print('The minimum CRD is:', min_crd)
print('The maximum CRD is:', max_crd)
```

### CLV calculation

To compute the CLV, we need information on Eurotel's WACC as well as the forecasted EBITDA rates. Let's import the Excel file that contains all the exhibits and extract the sheets that contain the EBITDA as a percentage of monthly revenue and the WACC.

We use the <a href="https://pandas.pydata.org/docs/reference/api/pandas.ExcelFile.parse.html">pd.ExcelFile.parse()</a> function that parses specified sheet(s) into a pandas DataFrame.

```python
%%capture

# download data from google drive to colab
!wget https://drive.google.com/uc?id=10o84GJ-_Qz3ts3RiFAQ28WJxiFRcL34R -O Eurotel_exhibit_data.xlsx
```

```python
exhibits = pd.ExcelFile('Eurotel_exhibit_data.xlsx')

month_wacc = exhibits.parse('WACC_month')
ebitda_perc = exhibits.parse(
    'Exhibit 2',index_col=0)

print(month_wacc)
print(ebitda_perc)
```

### CLV calculation

We can now calculate the CLV values per customer.

To do so, we use the <a href="https://numpy.org/numpy-financial/">numpy_financial</a> package.This package contains the <a href="https://numpy.org/numpy-financial/latest/npv.html">npf.npv()</a> function that returns the NPV (Net Present Value) of a cash flow.

```python
import numpy_financial as npf
```

<script src='https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.4/MathJax.js?config=default'></script> $$ \text{CLV}=\sum_{i=1}^{\text{CRD}}\dfrac{\text{EBITDA}_{{\text{month}}_i}}{(1+\text{WACC}_{\text{month}})^i}. $$




### CLV calculation

First, we construct two empty arrays that we use to store our calculations. Are you able to explain the shape of these arrays?

Next, we extract the WACC from the pandas Series and store the EBITDA data in the predefined array.

```python
array_ebitda_perc = np.zeros(max_crd)
clv = np.zeros(len(df))
wacc = month_wacc.iloc[0]['WACC_month']

array_ebitda_perc[0] = ebitda_perc['M1'].iloc[0]
array_ebitda_perc[1] = ebitda_perc['M2'].iloc[0]
array_ebitda_perc[2] = ebitda_perc['M3'].iloc[0]
array_ebitda_perc[3] = ebitda_perc['M4'].iloc[0]
array_ebitda_perc[4] = ebitda_perc['M5'].iloc[0]
array_ebitda_perc[5:max_crd] = ebitda_perc['M6-inf'].iloc[0]
```

### CLV calculation

Second, we construct arrays that contain the revenues per user and the customer relationship durations.

We focus on the first three customers to gather some insight. Afterwards we do the same calculation for all customers.

Let's take a look at the monthly revenue of the first 3 customers and their CRD.

```python
revenueperuser = np.array(df["RevenuePerUser"])
crd = np.array(df["CRD"]).astype(int)

clv3 = np.zeros(3)
revenueperuser3 = revenueperuser[0:3]
crd3 = crd[0:3]

print(revenueperuser3)
print(crd3)
```

### CLV calculation

Third, we multiply the EBITDA percentage (different for each month) with the monthly revenue.

To do so, we use the <a href="https://numpy.org/doc/2.1/reference/generated/numpy.outer.html">np.outer()</a> function. This function computes the outer product of two vectors: The result includes all possible combinations of the two vectors (ie. outer product).

We store this information in the two-dimensional array `ebitda3`. Can you  explain the shape of this array?

```python
ebitda3 = np.outer(revenueperuser3, array_ebitda_perc)

print(ebitda3)
print("\nThe shape of this array is:", ebitda3.shape)
```

### CLV calculation

We now have all the information to compute the CLV. To do so, we discount the EBITDA values using the monthly WACC and sum over the CRD value.

For instance for the second customer (CRD = 2):
    
<script src='https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.4/MathJax.js?config=default'></script> \begin{align} \text{CLV}&=\dfrac{30.616}{1+\text{WACC}_{\text{month}}} \\ &+ \dfrac{28.48}{(1+\text{WACC}_{\text{month}})^2}. \end{align}.








```python
# Cell left blank for students
for i, val in enumerate(crd3):

    clv3[i] = npf.npv(wacc,
                      ebitda3[i,0:val])/(1+wacc)

clv3
```

### CLV calculation

We now do the same calculations for all customers and include the CLV values in our DataFrame.

```python
matrix = np.outer(revenueperuser, array_ebitda_perc)

for i, months in enumerate(crd):
    cashflows = matrix[i, :months]
    clv[i] = npf.npv(wacc, cashflows) / (1 + wacc)

df["CLV"] = clv
df.head(5)
```

## Case question 3: Targeting customer and product segments

Questions:

* Based on the potential acquisition value for each product and customer segment, which segments are most attractive?  
* What are the five product and customer segments that Eurotel should target?

### Exercise: continued

To determine the potential acquisition value per segment, we first need to calculate the average CLV of each segment.

Let's do this for Eurotel's customer segments:


```python
avg_clv_customer_segments = df[["CustomerSegment", "CLV"]].groupby(["CustomerSegment"]).CLV.mean().reset_index()
avg_clv_customer_segments.rename(columns = {'CLV': 'Average CLV'}, inplace = True)

# Visualize the result
avg_clv_customer_segments
```

### Exercise: continued

Let's do the same for the product segments:

```python
avg_clv_product_segments = df[["ProductSegment", "CLV"]].groupby(["ProductSegment"]).CLV.mean().reset_index()
avg_clv_product_segments.rename(columns = {'CLV': 'Average CLV'}, inplace = True)

# Visualize the result
avg_clv_product_segments
```

### Exercise: continued

The next step is to determine the potential number of clients that could switch from our competitor Econnect.

To do so, we need:

* Exhibit 1, which contains the size of the product segments of Econnect.
* Exhibit 3, which contains the probabilities that capture the likelihood of a customers switching from Econnect to Eurotel.


```python
# Load the exhibits
switch_prob = exhibits.parse('Exhibit 3',
                             index_col=0)
size_Econnect_prod_seg = exhibits.parse('Exhibit 1',
                               index_col=0)
size_Econnect_prod_seg = size_Econnect_prod_seg.drop('TOTAL', axis=1)

# Determine the number of clients we can attract from Econnect
potential_per_seg = pd.DataFrame(switch_prob.values*size_Econnect_prod_seg.values,
                columns=switch_prob.columns,
                index=switch_prob.index)
potential_per_seg['Total Potential Acquired Customers'] \
           = potential_per_seg.sum(axis=1)
potential_per_seg
```

### Exercise: continued

We can now determine the potential acquisition value per product segment by multiplying the average CLV of each product segment with the total potential acquired customers per product segment.

```python
total_pot_acq_cust = potential_per_seg['Total Potential Acquired Customers']

per_segment_CLV = avg_clv_product_segments.set_index('ProductSegment')['Average CLV']
per_segment_CLV = pd.concat([per_segment_CLV,total_pot_acq_cust], axis = 1)
per_segment_CLV['Potential Acquisition Value'] = per_segment_CLV.product(axis=1)

# Visualize the result
per_segment_CLV
```

### Exercise: continued

Now that we know how many customers we can potentially attract in each of our product segments, we need one final piece of information:

The distribution of current Eurotel customers over the different product segments.

```python
distribution_of_customers = pd.crosstab(index=df['ProductSegment'],
                         columns=df['CustomerSegment']).apply(lambda r: r/r.sum(), axis=1)
distribution_of_customers = distribution_of_customers.rename(columns={"Adults-Children": "Adults-Children (%)",
                            "Adults-NoChildren": "Adults-NoChildren (%)",
                          "Senior": "Senior (%)","Youth": "Youth (%)"})

# Visualize the result
distribution_of_customers
```

### Exercise: continued

What segments does Eurotel's management want to prioritize?

```python
importance_scores = exhibits.parse('Exhibit 4',index_col=0)
importance_scores
```

### Exercise: continued

We can now determine wich product and customer segments have the highest value potential by multiplying the distribution of customers over the product segments with the potential acquisition value of each product segment.

Since we are expected to take into account the strategic importance scores when selecting which segments to target, let's include them into our final output.

```python
pot_acq_val= per_segment_CLV['Potential Acquisition Value']

seg_pot = distribution_of_customers.mul(pot_acq_val, axis=0).rename(
    columns={"Adults-Children (%)": "Adults-Children (Potential value)",
            "Adults-NoChildren (%)": "Adults-NoChildren (Potential value)",
             "Senior (%)": "Senior (Potential value)","Youth (%)": "Youth (Potential value)"})
final_output = pd.concat([distribution_of_customers, seg_pot], axis = 1)
final_output['Strategic Importance Score'] = importance_scores

# Visualize the final output
round(final_output, 2)
```

### Exercise: continued

Although we can use the previous output to select the 5 segments that should be prioritized, it's easier to do so based on a comprehensive visual. Let's construct a heatmap visualizing the segment potentials.

```python
ax = sns.heatmap(final_output[['Adults-Children (Potential value)','Adults-NoChildren (Potential value)',
                            'Senior (Potential value)', 'Youth (Potential value)']], linewidth=0.5,cmap="YlGnBu")
ax.set_facecolor('red')
ax.set_title('Heatmap Customer Segments')
plt.show()
```

* Mobile + Youth
* Web + Adults-NoChildren
