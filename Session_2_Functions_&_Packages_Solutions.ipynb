## 2.1 Functions

### What is a function

In Python, a function definition begins with the `def` keyword.

One can provide a list of parameters inside the round brackets `(.)`.

By using `return`, we can return a result to the function caller.

The function `add_number()` takes two numbers and adds them together.


```python
def add_number(x,y):
    sum = x + y
    return sum

add_number(2,4)
```

### Functions: A second example

It is possible to specify a default parameter value (`z=1` in the example).

When calling a function, use an argument (`2` and `4` in the example) or keyworded argument (`z = 3` in the example).

Functions make code more manageable and easier to understand.
   




```python
def add_numbers(x, y, z=1):
    sum = x + y + z
    return sum

print(add_numbers(2, 4))
print(add_numbers(2, 4,z=3))
```

#### Built-in functions: map & list

One important built-in function is the `map` function.

The `map()` function allows us to apply a given function on an iterable (you can iterate over it) and returns this iterable.

The `list()` function takes an iterable object as input and creates a new list.

This allows us to calculate the double of each element of a list.

```python
def addition(x):
    return x + x

num_list = [1, 2, 3]
result = map(addition, num_list)
print(list(result))
```

#### Lambda functions: a second example

A lambda function can have multiple inputs.

On the right, we calculate the sum of the elements of 2 different lists.

```python
num_list_1 = [1, 2, 3]
num_list_2 = [4, 5, 6]

result = map(lambda x, y: x + y,
         num_list_1, num_list_2)
print(list(result))
```

### Generating arrays


One can do statistical tests by using <a href="https://docs.scipy.org/doc/scipy/"> `Scipy`</a>.

```python
from scipy.stats import ttest_ind

female_length = np.random.normal(loc=168,
        scale=6,size=100000)
male_length = np.random.normal(loc=178,
        scale=6,size=100000)

res = ttest_ind(female_length,male_length)
print(res)
```

### Combining DataFrames

```python
cloth_1 = pd.DataFrame([{'Name':'Hat','Range':'Cheap'},
                        {'Name':'Shirt','Range':'Expensive'}]).set_index('Name')
cloth_2 = pd.DataFrame([{'Name':'Hat','Range':'Expensive'},
                        {'Name':'Dress','Range':'Expensive'}]).set_index('Name')
print(cloth_1)
print("\n")
print(cloth_2)
```

### Combining DataFrames

```python
print(pd.merge(cloth_1, cloth_2,
        how='left', on='Name'))
print("\n")
print(pd.merge(cloth_1, cloth_2,
        how='right', on='Name'))
```

```python
print(pd.merge(cloth_1, cloth_2,
        how='inner', on='Name'))
print("\n")
print(pd.merge(cloth_1, cloth_2,
        how='outer', on='Name'))
```

b) Melissa decides that once the total balance (principal plus interest) in the account reaches €14,000 or more, she will withdraw all the accumulated interest, leaving only the original €10,000 principal in the account.
**Display the exact amount of money that Melissa withdraws when the balance goes over the €14,000 and recalculate the amount of money in the account again until the end of year 10.** <br>

Tip: Use the compound interest formula for calculating the amount of money in the account after n years: <script src='https://cdnjs.cloudflare.com/ajax/libs/mathjax/2.7.4/MathJax.js?config=default'></script> $$ {p(1 + r)^n} $$

where <br>
<i>p</i> is the original amount invested (i.e., the principal) <br>
<i>r</i> is the annual interest rate <br>
<i>n</i> is the number of years

```python
# Exercise 2 a)
amount = 0
principal = 10000
rate = 0.05
for year in range(1, 11):
    amount = principal * (1 + rate) ** year
    print(f'{year:2}{amount:10.2f}')
```

```python
# Exercise 2 b)
principal = 10000
threshold = 14000
rate = 0.05
withdrawal = 0
for year in range(1, 11):
    if (year == 1):
        amount = principal
    amount = amount*(1+rate)
    print(f"Amount at the end of year {year:12} is {amount:10.2f}")
    if amount >= threshold:
        withdrawal = amount-principal
        amount = principal
        print(f"Amount withdrawn at the end of year {year:2} is {withdrawal:10.2f}")
```
